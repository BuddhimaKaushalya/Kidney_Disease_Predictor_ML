## Handling Misssing Values

```python
import numpy as np

# Load features and targets as pandas DataFrame
X = chronic_kidney_disease.data.features.copy()  
Y = chronic_kidney_disease.data.targets

# Replace "?" with NaN in the features DataFrame
X = X.replace("?", np.nan)  # Avoid inplace operation to bypass warning

# Combine features and target into a single DataFrame
df_combined = pd.concat([X, Y], axis=1)  

# Save to CSV file
df_combined.to_csv("chronic_kidney_disease.csv", index=False)
```

```python
# Count missing values in each feature
missing_values =X.isnull().sum()

# Display missing values count
print("\nMissing values in each feature:")
print(missing_values)
```

### Drop rows with missing values in columns where the count of missing data is small enough to be negligible

```python
X = X.dropna(subset=['pcc', 'ba' , 'htn', 'dm' , 'cad' , 'appet', 'pe' , 'ane'])

# Since rows are dropped from X, now we need to drop the corresponding rows from Y
Y = Y.loc[X.index]
print(X.shape)
print(Y.shape)
```

```python
# Count missing values in each feature
missing_values =X.isnull().sum()

# Display missing values count
print("\nMissing values in each feature:")
print(missing_values)
```

### Filling Missing Values with Mean or Mode

```python
# Define the imputation strategies for each column
impute_methods = {
    'su': 'mode', 'wbcc': 'mean', 'age': 'mean', 'rbcc': 'mean', 
    'al': 'mode', 'rbc': 'mode', 'pc': 'mode', 'bgr': 'mean', 
    'pot': 'mean', 'hemo': 'mean', 'pcv': 'mean', 'sod': 'mean', 
    'bp': 'mean', 'bu': 'mean', 'sc': 'mean', 'sg': 'mode'
}

# Handle missing values using the specified methods
for column, method in impute_methods.items():
    if method == 'mean':
        X[column] = X[column].fillna(X[column].mean())
    elif method == 'mode':
        X[column] = X[column].fillna(X[column].mode()[0])

# Display the updated DataFrame
print(X)
```

```python
# Count missing values in each feature
missing_values =X.isnull().sum()

# Display missing values count
print("\nMissing values in each feature:")
print(missing_values)
```

### Now the data set has no missing values

###  Convert  to binary values

```python
# Ensure the column is converted to strings
Y['class'] = Y['class'].astype(str)

# Remove leading and trailing whitespace
Y['class'] = Y['class'].str.strip()

# Replace unwanted characters
Y['class'] = Y['class'].replace({'ckd\t': 'ckd', 'ckd\n': 'ckd', 'notckd\t': 'notckd', 'notckd\n': 'notckd'})

# Map the cleaned values to binary integers
Y['class'] = Y['class'].map({'ckd': 1, 'notckd': 0})


# Ensure the column is converted to strings
X['dm'] = X['dm'].astype(str)

# Remove leading and trailing whitespace
X['dm'] = X['dm'].str.strip()

# Replace unwanted characterss
X['dm'] = X['dm'].replace({'yes\t': 'yes', 'yes\n': 'yes', 'no\t': 'no', 'no\n': 'no'})

# Map the cleaned values to binary integers
X['dm'] = X['dm'].map({'yes': 1, 'no': 0})

X['htn'] = X['htn'].map({'yes': 1, 'no': 0})
X['cad'] = X['cad'].map({'yes': 1, 'no': 0})
X['pe'] = X['pe'].map({'yes': 1, 'no': 0})
X['appet'] = X['appet'].map({'good': 1, 'poor': 0})
X['ane'] = X['ane'].map({'yes': 1, 'no': 0})
X['rbc'] = X['rbc'].map({'normal': 0, 'abnormal': 1})
X['pc'] = X['pc'].map({'normal': 0, 'abnormal': 1})
X['pcc'] = X['pcc'].map({'notpresent': 0, 'present': 1})
X['ba'] = X['ba'].map({'notpresent': 0, 'present': 1})

```

### Load and Process the dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler  # Combine the updated X (features) and Y (target) back into a single DataFrame
df_combined_updated = pd.concat([X, Y], axis=1)

# Save the combined DataFrame to a CSV file
df_combined_updated.to_csv("chronic_kidney_disease_updated.csv", index=False)
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from sklearn.neural_network import MLPClassifier
import pickle
import numpy as np

# Load the dataset
df = pd.read_csv('chronic_kidney_disease_updated.csv')

# Encode categorical variables
label_encoder = LabelEncoder()
df['class'] = label_encoder.fit_transform(df['class'])

# Split the dataset into features (X) and target (y)
X = df.drop('class', axis=1)
y = df['class']

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Define and Train Individual Models


```python
# Define individual models
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=500, random_state=42)

# Train each model
print("Training individual models...")
rf_model.fit(X_train, y_train)
gb_model.fit(X_train, y_train)
mlp_model.fit(X_train, y_train)

# Evaluate individual models
print("\nEvaluating individual models...")
models = [rf_model, gb_model, mlp_model]
for model in models:
    y_pred = model.predict(X_test)
    print(f"\nModel: {model.__class__.__name__}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(classification_report(y_test, y_pred))
```

### Create and Train the Ensemble Model


```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create and train the ensemble model
voting_model = VotingClassifier(estimators=[
    ('rf', rf_model),
    ('gb', gb_model),
    ('mlp', mlp_model)
], voting='hard')
print("\nTraining ensemble model...")
voting_model.fit(X_train, y_train)

# Evaluate the ensemble model
print("\nEvaluating ensemble model...")
voting_pred = voting_model.predict(X_test)

# Calculate accuracy and confusion matrix for the ensemble model
ensemble_accuracy = accuracy_score(y_test, voting_pred)
ensemble_conf_matrix = confusion_matrix(y_test, voting_pred)

print(f"Ensemble Model Accuracy: {ensemble_accuracy}")
print("Ensemble Model Confusion Matrix:")
print(ensemble_conf_matrix)

# Visualize the confusion matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(ensemble_conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=label_encoder.classes_, yticklabels=label_encoder.classes_)
plt.title("Confusion Matrix for Ensemble Model")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```

### Simulate Real-Time Predictions

```python
print("\nSimulating Real-Time Predictions...")
for i in range(5):  # Adjust the range to simulate more samples
    print(f"\nPredicting Sample {i + 1}...")
    test_sample = X_test[i:i + 1]

    # Predict with individual models
    rf_pred = rf_model.predict(test_sample)[0]
    gb_pred = gb_model.predict(test_sample)[0]
    mlp_pred = mlp_model.predict(test_sample)[0]

    # Predict with the ensemble model
    voting_pred = voting_model.predict(test_sample)[0]

    print(f"Random Forest Prediction: {rf_pred}")
    print(f"Gradient Boosting Prediction: {gb_pred}")
    print(f"MLP Prediction: {mlp_pred}")
    print(f"Final Ensemble Prediction (Majority Vote): {voting_pred}")
```
